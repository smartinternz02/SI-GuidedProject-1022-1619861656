```python
import seaborn as sns
```

```python
iris=sns.load_dataset("iris")
```

```python
iris
```

```python
sns.scatterplot(x="sepal_length",y="sepal_width",data=iris,color="green")
```

```python
sns.scatterplot(x='petal_length',y='petal_width',data=iris,color='red')
```

```python
sns.lineplot(x="sepal_length",y="petal_width",data=iris,color="blue")
```

```python
sns.pairplot(iris)
```

```python
sns.barplot(x='sepal_width',y='species',data=iris)
```

```python
sns.barplot(x='species',y='petal_length',data=iris)
```

```python
sns.boxplot(x='species',y='petal_width',data=iris)
```

```python
sns.boxplot(x='species',y='sepal_width',data=iris)
```

```python
sns.distplot(iris['sepal_length'])   #We cannot have distplot for string type of data
```

```python
sns.catplot(x='species',y='petal_width',data=iris)
```

```python
sns.heatmap(iris.corr(),annot=True)
```
